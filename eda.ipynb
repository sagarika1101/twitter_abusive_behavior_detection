```python
def n_grams(df, class_name, min_gram, max_gram):
    bow_vectorizer = CountVectorizer(input='content',
                                 analyzer='word',
                                 strip_accents='ascii',
                                 ngram_range=(min_gram,max_gram),
                                 stop_words='english',
                                 min_df=10)
    df_class = df[df['label'] == class_name]
    X = bow_vectorizer.fit_transform(df_class['cleaned'])
    feature_names = bow_vectorizer.get_feature_names()
    counts = np.array(X.sum(axis=0)).squeeze()
    new_df = pd.DataFrame(counts)
    new_df.columns = ['counts']
    new_df["features"] = feature_names
    return new_df
```

```python
df_s = n_grams(df, "abusive", 2, 5)
df_s.nlargest(50, "counts")
```
